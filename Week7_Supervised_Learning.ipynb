**Supervised Learning** - *RECAP*

Supervised learning is the most common subfield of Machine Learning. At its core, it is about teaching by example.

In supervised learning, you provide an algorithm with a dataset that contains both the input data and the correct answers (labels). The goal is for the model to learn the relationship between the two so it can accurately predict the label for new, unseen data.

It is called "supervised" because the process of an algorithm learning from the training dataset can be thought of as a teacher supervising the learning process. We know the correct answers beforehand, and the algorithm iteratively makes predictions on the training data and is corrected by the "teacher" until it achieves an acceptable level of performance.

The supervised learning in general focuses on two different types of problems:

Classification is used when the output is categorical. This means the model assigns data points into discrete groups or "labels."
