Numpy, short for Numerical Python, is one of the leading packages for numerical computing in Python. Numpy's array objects are one of the standard interfaces for data exchange.

Numpy includes:
- ndarray an efficient multi dimensional array providing fast array-oriented arithmetic operations.
- Mathematical functions for fast operations on entire arrays of data without having to write loops.
- Tools for reading / writing array data to disk and. working with memory-mapped files.
- Linear algebra, random number generation, and. Fourier transform capabilities.


While Numpy itself does not provide modeling and scientific functionality, understanding NumPy arrays and array-oriented computing will help you use tools with array computing semantics.

##Series

A Series is a one-dimensional data structure consisting of labeled data. The labels are referred to as the index. The data itself can consist of numbers, strings, or other Python objects. Series can be created using lists, arrays, or dictionaries which are inbuilt data structures for Python.








**Let's create a series object from data we loaded by using pd.read_csv()** The read_csv function imports data to the Python environment in **DataFrame** structure, which we will discuss later...

So we will create a series object from a dataframe object.

```
nufus_series = pd.Series(Data['Nufus'].values, index=Data['Mahalle_Adi'])
(nufus_series)
```

Series objects are fixed-length, ordered dictionaries; they are mapping of index—values to data values. Using Numpy like functions one can do operations such as filtering with Boolean array, scalar multiplication or applying. math functions

**Let's filter neighbourhoods with exactly 10000 population**

```
nufus_series[nufus_series == 10000]
```

** Filter neighbourhoods with more than 20000 inhabitnats.**

```
nufus_series[nufus_series > 20000]
```

**Or we can conduct lograithmic transformation to the number of inhabitants in. each neighbourhood**

```
import numpy as np
np.log(nufus_series)
```
